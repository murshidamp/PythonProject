```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Data = pd.read_excel('ABCemployees.xlsx')
```

```python
print(Data.head())
```

```python
Data['Height']= np.random.randint(150,181,size=(len(Data)))
print(Data.head())
data=pd.DataFrame(Data)
data.to_excel('ABCEmployees.xlsx')
```

```python
team_count=Data['Team'].value_counts()
split_percentage=(team_count/len(Data)*100)
print('Percentage splitting with respect to total employees in different team: ','\n',split_percentage)
plt.figure(figsize=(8,8))
plt.pie(team_count,labels=team_count.index,autopct='%1.1f%%')
plt.title('Percentage splitting with respect to total employees in different team')
plt.show()

```

```python
position_count=Data['Position'].value_counts()
print('Employees segregated with respect to position:','\n',position_count)
plt.figure(figsize=(6,4))
plt.bar(position_count.index,position_count)
plt.title('Employees segregated with respect to position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
plt.hist(Data['Age'],bins=[20,25,30,35,40,45],edgecolor='red')
plt.xlabel('Age')
plt.ylabel('Employees')
plt.title('Age distribution of employees')
plt.show()
```

```python
high_salary=Data.groupby(['Team','Position'])['Salary'].sum().idxmax()
print('Team and Position with highest spending in terms of salary:','\n',high_salary)
```

```python
correlation=Data['Age'].corr(Data['Salary'])
print('Correlation_coefficient between age and salary: ',correlation)
sns.scatterplot(x='Age',y='Salary',data=Data)
plt.show()

```

```python
Age_Salary=Data[["Age","Salary"]]
corr_coefficient=Age_Salary.corr()
corr_coefficient
print('Correlation Coefficient between Age and Salary')
sns.heatmap(corr_coefficient,annot=True,cmap="coolwarm")
```
